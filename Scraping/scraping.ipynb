```python
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError

from bs4 import BeautifulSoup
#ELSE DEPOIS DE TRY-EXCEPT
#Se você deseja continuar o código, mas ele precisa ter outro try except para funcionar.

def getTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try: 
        bs = BeautifulSoup(html.read(), 'html.parser')
        title = bs.h1
    except AttributeError as e:
        return None
    return title

title = getTitle("http://www.faeterj-petropolis.edu.br/site/avisos/")
if title == None:
    print("Título não encontrado.")
else:
    print(title)
```
